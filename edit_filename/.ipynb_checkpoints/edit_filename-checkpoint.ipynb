```python
import pandas as pd

def north_img_num(dir):
    if dir < 50 or 300 < dir:
        return 4
    elif 50 < dir < 150:
        return 3
    else:
        return 2

data = pd.read_csv("all_.csv", index_col=0)
df_direction = pd.DataFrame(index=data.index, columns=['north', 'west', 'south', 'east'])
df_direction['north'] = data['compass_direction'].map(north_img_num)
df_direction['west'] = df_direction['north'].map(lambda x: (x + 2) % 4 + 1)
df_direction['south'] = df_direction['north'].map(lambda x: (x + 1) % 4 + 1)
df_direction['east'] = df_direction['north'].map(lambda x: (x) % 4 + 1)
df_group_1_with_dir_num = data.join(df_direction)

group_1 = df_direction[df_direction['north']==4]
group_2 = df_direction[df_direction['north']==3]
group_3 = df_direction[df_direction['north']==2]

group_1 = list(group_1.index)
group_2 = list(group_2.index)
group_3 = list(group_3.index)
```

```python
import os
# ここでファイルを保存しているディレクトリを指定してください
os.chdir("D:/worldmodeldata")

import glob
files = glob.glob("*.jpg")

# 0 indexにする為に-3078する(1 + 3077)

os.getcwd()
```

```python
def node_int_to_string(node):
    node = str(node)
    temp = len(node)
    fill = 6 - temp
    fill_text = "0" * fill
    node = fill_text + node
    return node


for i in files:
    rokuketa = int(i[:6])
    rokuketa -= 3078
    
    if rokuketa < 0:
        os.remove(i)
        continue
    elif rokuketa >= 1325:
        # 少し多めにとってます
        os.remove(i)
        continue
    
    if rokuketa in group_1:
        if i[7] == "1":
            direction = "0"
        elif i[7] == "4":
            direction = "1"
        elif i[7] == "3":
            direction = "2"
        elif i[7] == "2":
            direction = "3"
        elif i[7] == "0" or i[7] == "5":
            os.remove(i)
            continue
        else:
            print(f'unexpected error!!! stopped at {i}')
            break
        
    elif rokuketa in group_2:
        if i[7] == "4":
            direction = "0"
        elif i[7] == "3":
            direction = "1"
        elif i[7] == "2":
            direction = "2"
        elif i[7] == "1":
            direction = "3"
        elif i[7] == "0" or i[7] == "5":
            os.remove(i)
            continue
        else:
            print(f'unexpected error!!! stopped at {i}')
            break
    
    else:
        # 3044 (group_3)は面倒だったのでやってません......
        continue
        
    before_ = node_int_to_string(rokuketa)
    file_name = before_ + "_" + direction + ".jpg"
    
    os.rename(i, file_name)
    print(file_name)
```
